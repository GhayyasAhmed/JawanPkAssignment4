# Program 1

```python
#Make a calculator using Python with addition , subtraction ,multiplication ,division and power.


print("Enter what operation you want to perform: ")
operators = ["+", "-", "*", "/", "**"]
operator = input("Addition(+), Subtraction(-), Multiplication(*), Division(/), Power(**): ")

if operator in operators:
    num1 = float(input("Enter number 1: "))

    num2 = float(input("Enter number 2: "))

    if operator == "+":
        res = num1 + num2
        print(num1,"+",num2,"=",res)

    elif operator == "-":
        res = num1 - num2
        print(num1,"-",num2,"=",res)

    elif operator == "*":
        res = num1 * num2
        print(num1,"*",num2,"=",res)

    elif operator == "/":
        res = num1 / num2
        print(num1,"/",num2,"=",res)

    elif operator == "**":
        res = num1 ** num2
        print(num1,"**",num2,"=",res)

else:
    print("Invalid operator")
```

# Program 2

```python
#Write a program to check if there is any numeric value in list using for loop.

lst = ["Ghayyas", "Ahmed", "JawanPk", 14, "Student", 2021]
flag = True
for i in lst:
    if type(i) == int or type(i) == float:
        print("Yes there is atleast one number in the list")
        flag = False
        break

if flag:
    print("There isn't any number in the list")
```

# Program 4

```python
# Write a Python program to sum all the numeric items in a dictionary.

dict2 = {"Name": "Ghayyas", 1: 1, 2:2, 3:3.5}
numericSum = 0
for i in dict2.items():
    for j in i:
        if type(j) == int or type(j) == float:
            # Add all keys and values of type int and float
            numericSum += j

print("Sum of numeric items in dict =", numericSum)
```

# Program 5

```python
# Write a program to identify duplicate values from list.


lst = ["Ghayyas", "Ahmed", "JawanPk", 7, "Ghayyas", 7, 31, 7, "JawanPk"]
checked = []

for i in lst:
    if lst.count(i) > 1:
        if i not in checked:
            print(i,"comes",lst.count(i),"times in the list")
            checked.append(i)
```

# Program 6

```python
#Write a Python script to check if a given key already exists in a dictionary

dict3 = {"FirstName": "Ghayyas", "LastName": "Ahmed"}
newKey = input("Enter a key to insert in the dict: ")
if newKey not in dict3.keys():
    newValue = input("Enter value for this key: ")
    dict3[newKey] = newValue
    print("Dictionary updated")
    print(dict3)
else:
    print("Key already exists. Enter a unique key")
```
